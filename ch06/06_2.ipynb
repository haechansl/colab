```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/gdrive')
```

```python
img= cv2.imread('/content/gdrive/My Drive/Colab IP/Images/lena.jpg')
```

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
output_img = cv2.Canny(gray_img,100,200)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.title('Image1')
plt.imshow(gray_img,cmap='gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.title('Canny Edge Detection')
plt.imshow(output_img,cmap='gray')
plt.axis("off")
plt.show()
```

```python
img= cv2.imread('/content/gdrive/My Drive/Colab IP/Images/Lena-Gaussian-noise1.jpg')
```

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
blur = cv2.GaussianBlur(gray_img,(3,3),1)
laplacian = cv2.Laplacian(blur,-1,1)
output_img=laplacian/laplacian.max()
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.title('Image1')
plt.imshow(gray_img,cmap='gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.title('Gaussian-Laplacian Edge Detection')
plt.imshow(output_img,cmap='gray')
plt.axis("off")
plt.show()
```
